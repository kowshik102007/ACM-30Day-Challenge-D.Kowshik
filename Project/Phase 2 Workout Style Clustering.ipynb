```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
```

# Loading the csv file

```python
file = ('workout_fitness_tracker_data.csv')
df = pd.read_csv(file)
df.head()
```

```python
df= df.drop('User ID', axis=1)
```

# Encoding categorical columns using OHE

```python
df = pd.get_dummies(df,columns = ['Gender', 'Workout Type'])

le = LabelEncoder()
df['Workout Intensity'] = le.fit_transform(df['Workout Intensity'])

before_map = {
    'Tired': 0,
    'Stressed' : 0,
    'Neutral': 1,
    'Happy': 2
}
after_map = {
    'Fatigued': 0,
    'Neutral': 1,
    'Energized': 2
}

df['Mood Before Workout'] = df['Mood Before Workout'].map(before_map)
df['Mood After Workout'] = df['Mood After Workout'].map(after_map)

df.head() 
```

# Standardization and PCA reduction

```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)

pca = PCA(n_components=2)
df_pca = pca.fit_transform(df_scaled)
```

#  k-Means clustering to form 3 groups

```python
df_clusters = pd.DataFrame(df_pca, columns=['PCA1', 'PCA2'])

kmeans = KMeans(n_clusters=3, random_state=2)
df_clusters['cluster'] = kmeans.fit_predict(df_pca)

silhouette_score = silhouette_score(df_pca, df_clusters['cluster'])
print("Silhouette Score :", round(silhouette_score, 3))
```

# Plot visualisation of clustering

```python
plt.figure(figsize=(6, 5))
sns.scatterplot(x=df_clusters['PCA1'], y=df_clusters['PCA2'], hue=df_clusters['cluster'], palette='Set1', s=50)
plt.title('2D PCA Plot of Workout Style Clustering ')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster')
plt.show()
```
